Unsere trliste ist ja die Liste aller Geschäfte. Jetzt gehen wir in den ersten Eintrag [0] der trliste und steuern alle Elemente nacheinander an. 

```python
td_list = trlist[0].find_all('td')
url = td_list[0].find('a')['href'] #das ist die URL, die wir speziell auslesen müssen.
geschäft = td_list[1].text #entspricht Geschäft, hier wie bei den folgenden können wir einfach den Text auslesen.
stand = td_list[2].text
zuständig = td_list[3].text
geschäftsart = td_list[4].text
status = td_list[5].text

mini_dict = {'url':url,
          'geschäft': geschäft,
          'stand': stand,
          'zuständig': zuständig,
          'geschäftsart': geschäftsart,
          'status' : status}
        
```

```python
mini_dict
```

Jetzt bauen wir einen Loop und machen das ganze für die ganze Seite (page 1 der Geschäftsliste im Web). 

```python
seite01 = []
for element in trlist:
    td_list = element.find_all('td')
    url = td_list[0].find('a')['href'] 
    geschäft = td_list[1].text
    stand = td_list[2].text
    zuständig = td_list[3].text
    geschäftsart = td_list[4].text
    status = td_list[5].text

    mini_dict = {'url':url,
          'geschäft': geschäft,
          'stand': stand,
          'zuständig': zuständig,
          'geschäftsart': geschäftsart,
          'status' : status}
    
    seite01.append(mini_dict)
```

```python
import pandas as pd
pd.DataFrame(seite01)
```

Eigentlich sind wir noch nicht fertig, wir haben ja nur die Daten der ersten Seite heruntergeladen. Es sind aber 40 Seiten mit Geschäften. 

```python
adresse = 'https://kr-geschaefte.zug.ch/gast/geschaefte?commit=Filtern&geschaeft_filter%5Babgeschlossen_bis%5D=&geschaeft_filter%5Babgeschlossen_von%5D=&geschaeft_filter%5Barten_refs%5D%5B%5D=&geschaeft_filter%5Beingereicht_bis%5D=&geschaeft_filter%5Beingereicht_von%5D=&geschaeft_filter%5Bfrist_bis%5D=&geschaeft_filter%5Bhistorische_staende_refs%5D%5B%5D=&geschaeft_filter%5Bkommissionen_refs%5D%5B%5D=&geschaeft_filter%5Bstaende_refs%5D%5B%5D=&geschaeft_filter%5Bstatus_ids%5D%5B%5D=haengig&geschaeft_filter%5Bstatus_ids%5D%5B%5D=abgeschlossen&geschaeft_filter%5Bstatus_ids%5D%5B%5D=&geschaeft_filter%5Btitel%5D=&geschaeft_filter%5Bzustaendig_refs%5D%5B%5D=&page='
alleseiten = []

for seite in range(1,41):
    r = r = requests.get(adresse + str(seite))
    soup = BeautifulSoup(r.text, 'html.parser')
    
    trlist = soup.find_all('tr')[2:]


    for element in trlist:
        td_list = element.find_all('td')
        url = td_list[0].find('a')['href'] 
        geschäft = td_list[1].text
        stand = td_list[2].text
        zuständig = td_list[3].text
        geschäftsart = td_list[4].text
        status = td_list[5].text

        mini_dict = {'url':url,
              'geschäft': geschäft,
              'stand': stand,
              'zuständig': zuständig,
              'geschäftsart': geschäftsart,
              'status' : status}
    
        alleseiten.append(mini_dict)
```
