Und jetzt noch für die 35 Seiten. Dazu definieren wir eine Funktion mit dem obigen Code. 

```python
def extract_bakeries(elem):
    
    b_list = []
    zeilen = h.find_all('tr', {'class':'ng-scope'})

    for zeile in zeilen:
        
        name = zeile.find('div').text
        u_id = zeile.find_all('div')[2].text
        sitz = zeile.find_all('div')[4].text
        kanton = zeile.find_all('div')[5].text
        url = zeile.find('div').find_next('div').find('a')['href']

        minidict = {'Name': name,
                   'UID': u_id,
                   'Sitz': sitz,
                   'Kanton': kanton,
                   'URL': url}

        b_list.append(minidict)
    
    return b_list  #das gibt die Liste zurück (bzw. aus) für eine weitere Verwendung.
```

Nachher wird es relativ simpel, weil wir einfach die Funktion *extract_bakeries* aufrufen können. Wir müssen jetzt einen Loop bauen, um alle 35 htm auszulesen und in ein DataFrame zu speichern. 

```python
lst = os.listdir('zefix_result')
final_list = []

for filename in lst: # das ist unsere Liste aller Files im Ordner (mit os erstellt)
    file = open('zefix_result/' + filename, 'r')
    text = file.read()
    h = BeautifulSoup(text, 'html.parser')

    final_list = final_list + extract_bakeries(h)
```

```python
pd.DataFrame(final_list)
```
