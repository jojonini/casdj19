Die URL: https://records.nhl.com/records/skater-records/seasons-and-games/most-seasons

```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import Select
import time
import pandas as pd
from bs4 import BeautifulSoup
```

```python
#Seite abfrufen
driver = webdriver.Chrome(executable_path='/usr/local/bin/chromedriver')
driver.get('https://records.nhl.com/records/skater-records/seasons-and-games/most-seasons')

#Seite abspeichern
page = driver.page_source.encode('utf-8')
#print(page)
with open("nhl_records.htm", "wb+") as file:
    file.write(page)
    file.close()
    
```

```python
#Seite öffnen und in bs einlesen
file = open('nhl_records.htm', 'r')
text = file.read()
h = BeautifulSoup(text, 'html.parser')
```
