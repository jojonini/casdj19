### Was ist Pandas?
Kurz gesagt: Pandas ist Excel für Python.

Das Modul ermöglicht ähnliche Operationen wie ein Tabellenkalkulationsprogramm, aber
- in Logbuchform statt als Tabellenblatt
- mit reproduzierbarem Code
- viel mehr Funktionalität
- kombinierbar mit anderen Modulen

Ohne Pandas (fast) kein Datenjournalismus mit Python.

## In Pandas...

Statt in Excel öffnen wir das File nun in Pandas und versuchen, Antworten auf einige unserer Fragen zu erhalten.

### File öffnen

```python
path = "dataprojects/powerplants/global_power_plant_database.csv"
```

```python
df = pd.read_csv(path) #df als Variable für Datafile (unsere Datenbank)
```

```python
df.head(10)
```

```python
df.sort_values("capacity_mw",ascending=False)
```

### Inhalt anzeigen

**Übersicht über die Spalten**

```python
df.dtypes
```

**Visuelle Repräsentation**

```python
df.head(10)
```

```python
df.groupby("country_long")["capacity_mw"].sum().sort_values(ascending=False)
```

Jetzt, wie ist es mit Atomkraftwerken?

```python
df[df['primary_fuel'] == 'Nuclear']
```

```python
df.groupby('country_long')['capacity_mw'].sum().sort_values(ascending=False).head(10).plot(kind='barh')
```

Learning: ein einfaches Tool, um sich über grössere Datenmengen einen Überblick zu verschaffen. Und man verändert nicht wie in Excel die Daten oder schafft sich neue Tabellen und Grafiken, ohne, dass man weiss was getan wurde. Hier im Notebook mit Pandas hat man direkt eine Dokumentation, was mit den Daten gemacht wurde.
