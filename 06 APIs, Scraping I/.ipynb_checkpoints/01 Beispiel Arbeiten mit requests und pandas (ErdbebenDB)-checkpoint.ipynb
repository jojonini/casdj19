Jetzt machen wir die Abfrage für requests und speichern die Resultate in die Liste (new_list):

```python
url_anfang = "https://earthquake.usgs.gov/fdsnws/event/1/count?starttime=2019-09-0"
url_mitte = "&endtime=2019-09-0"
new_list = []
for tag in range(1,5):
    print(url_anfang + str(tag) + url_mitte + str(tag + 1))
    abfrage = requests.get(url_anfang + str(tag) + url_mitte + str(tag + 1))
    print(abfrage.text)
    new_list.append(abfrage.text)
    
    
```

```python
print(new_list)
```

```python
for line in new_list:
    print(line)
```

Jetzt gehen wir tiefer rein. Wir wollen jetzt Daten rausziehen, die uns interessieren. Zum Beispiel place, magnitude und time. Zuerst zum Testen mit print, um zu schauen, ob dabei rauskommt, was wir suchen. 

```python
for beben in erdbeben:
    print(beben["properties"]["mag"])
    print(beben["properties"]["place"])
    print(beben["properties"]["time"])
```

Jetzt bauen wir damit eine neue Liste:

```python
eigeneliste = []
for beben in erdbeben:
    magnitude = beben["properties"]["mag"]
    place = beben["properties"]["place"]
    time = beben["properties"]["time"]
    mini_dict = {"Magnitude: ":magnitude, "Place: ":place, "Time: ":time}
    
    eigeneliste.append(mini_dict)
```

```python
eigeneliste
```

Und jetzt visualisieren wir diese Liste mit Pandas:

```python
pd.DataFrame(eigeneliste)
```

Und jetzt speichern wir das Ergebnis noch ab, als .csv-Datei:

```python
pd.DataFrame(eigeneliste).to_csv("erdbebenliste.csv")
```

```python
import time #eine andere Bibliothek als datedime, etwas simpler.

eigeneliste = []

for beben in erdbeben:
    magnitude = beben["properties"]["mag"]
    place = beben["properties"]["place"]
    zeit = beben["properties"]["time"] #Achtung: time it jetzt auch die Funktion der Bibliothek time!
    zeit = time.ctime(int(zeit))
    mini_dict = {"Magnitude: ":magnitude, "Place: ":place, "Time: ":zeit}
    
    eigeneliste.append(mini_dict)
```

```python
pd.DataFrame(eigeneliste)
```
