# Komplizierte JSONs auslesen

```python
import pandas as pd
import numpy as np
import glob
import json
%matplotlib inline
```

```python
path = 'psyreg_output/'
filenames = glob.glob(path + '*.json')
```

```python
len(filenames)
```

```python
data = []
with open(filenames[0]) as json_data:
    data = json.load(json_data)
data
```

Eine der verschachtelten Listen von Dicts ist zum Beispiel das Feld 'bewilligungErteilt'. 

```python
type(data["bewilligungenErteilt"]) == list
```

```python
output = {}
for k,v in data.items(): #k für Key, v = value #.items holt k und v
    if type(v) == list: #wenn bei Value eine Liste kommt, dann bearbeite das wie folgt:
        dummy_keys = []
        for i in range(0,3): #wir gehen mal davon aus, dass nicht mehr als 4 Listen enthalten sind. Muss evlt. angepasst werden
            try:
                element = v[i]
                if len(dummy_keys) > 0:
                    dummy_keys = []
                if type(element) == int:
                    #print("been here int")
                    output[k] = element
                else:
                    #print("been here")
                    for k_inside, v_inside in element.items():
                        dummy_keys.append(k_inside)
                        #print(dummy_keys)
                        if type(v_inside) == type([]):
                            for inner_element in v_inside:
                                for k_inside_inside,v_inside_inside in inner_element.items():
                                    dummy_keys.append(k_inside+"-"+k_inside_inside)
                                    #print(dummy_keys)
                                    output[k+"-"+str(i)+"-"+k_inside+"-"+k_inside_inside] = v_inside_inside
                        else:
                            output[k+"-"+str(i)+"-"+k_inside] = v_inside
            except:
                #print(dummy_keys)
                for key in dummy_keys:
                    output[k+"-"+str(i)+"-"+ key] = ''
    else:
        #print("outer")
        output[k] = v
pd.DataFrame(output,index=[0]).columns
```

Bei Abschlüsse fehlt jetzt 55001. Der Grund: es wurde beim Erstellen des Minidicts überschrieben. 
Das heisst, wir müssen neue Keys generieren. Das Problem: bei gewissen Einträgen werden die Kolonnen der Tabelle nicht mehr passen. 


```python
liste = []
minidict = {}

for k,v in data.items():
    if type(v) == list:
        for i in range(0,3):
            try:
                print('juhu, hier ist was')
                minidict[k+str(i)] = v[i] #hier erstellen wir neue Keys
            except: 
                print('Mist, Element gibt es nicht')
                minidict[k+str(i)] = v[i]
        print('Das war jetzt gerade die Extrawurst (Liste)')
    else:
        print('Alles gut')
        minidict[k] = v
liste.append(minidict)
pd.DataFrame.from_dict(liste)

```
